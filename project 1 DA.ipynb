```python
import matplotlib.pyplot as plt
percent_ed=data['EdLevel'].value_counts(normalize=True)*100
percent_ed=percent_ed.sort_values()

percent_ed.plot(kind='barh')
plt.title('DISTRIBUTION OF eDUCATION LEVEL')
plt.xlabel('percentage')
plt.ylabel("education level")
```

```python
learn_methods=data['LearnCode'].str.split(";").explode()
method_counts=learn_methods.value_counts()
#.explode burst the list into its options

method_percentage=(method_counts /89184)*100
method_percentage
```

```python
method_percentage.plot(kind='barh')
plt.title("learning methods")
plt.ylabel('leanr code methods')
plt.xlabel('percentage')
for index,value in enumerate(method_percentage):
    plt.text(value,index,f'{value:.2f}%',va='center')
plt.show()
```

```python
online_counts=data["LearnCodeOnline"].str.split(";").explode().value_counts()
online_percent=(online_counts/89184)*100

online_percent
```

```python
online_percent.plot(kind='barh')
plt.title("online learning methods")
plt.ylabel('leanr code online methods')
plt.xlabel('percentage')
for index,value in enumerate(online_percent):
    plt.text(value,index,f'{value:.2f}%',va='center')
plt.show()
```
